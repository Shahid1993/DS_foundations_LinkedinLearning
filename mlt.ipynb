```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.linspace(-6,6,100)
ys = np.sinc(xs)
```

```python
plt.plot(xs,ys)
```

### Adding legends to plots

```python
plt.plot(xs, ys, label=r'$\frac{sin(x)}{x}$')
plt.legend()
```

### Multiple plots stacked

```python
plt.plot(xs, ys, label=r'$\frac{sin(x)}{x}$')
plt.plot(xs, np.sin(xs), label='sin(x)')
plt.legend()
```

### Multiple plots side by side

```python
plt.subplot(2,1,1)
plt.plot(xs, np.sin(xs), label='sin')
plt.legend()
plt.subplot(2,1,2)
plt.plot(xs, -xs**2, label=r'${-x}^2$')
plt.legend()
```
