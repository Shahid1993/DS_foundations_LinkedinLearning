### Edge Detection using `Canny Algorithm`

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
```

To display grayscale image correctly, we need to tell `plt` that the image is grayscale

```python
plt.imshow(gray, cmap=plt.cm.gray)
```

```python
edges = cv2.Canny(gray, 200, 300)
plt.imshow(edges, cmap=plt.cm.gray)
```

## Facial Detection

```python
img = cv2.imread('child.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
import sys
fname = '{}/Library/etc/haarcascades/haarcascade_frontalface_default.xml'.format(sys.prefix)
model = cv2.CascadeClassifier(fname)
model
```

```python
faces = model.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(500, 500))
```

```python
faces
```

```python
img = cv2.imread('child.jpg')
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 255), 50)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```
